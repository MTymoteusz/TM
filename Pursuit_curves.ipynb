# Notka historyczna 

Historia krzywych pościgowych rozpoczęła się od badań nad ruchem w XVIII wieku, kiedy to matematycy tacy jak Isaac Newton i Leonhard Euler zaczęli rozwijać rachunek różniczkowy i mechanikę analityczną. W XIX wieku zaczęto formalnie opisywać trajektorie ścigających obiektów, a rozwój tej teorii nabrał tempa w XX wieku, zwłaszcza w kontekście technologii rakietowych i wojskowych. Krzywe pościgowe były ogólnie rozważane przez francuskiego naukowca Pierre'a Bouguera w 1732 roku. Jednak termin "krzywa pościgowa" został po raz pierwszy zdefiniowany przez George'a Boole'a w jego "Traktacie o równaniach różniczkowych" w 1859 roku. [[1]](#bibliografia)

Krzywe pościgowe znajdują szerokie zastosowanie w wielu dziedzinach nauki, technologii i praktyki. Tego typu modele pomagają zrozumieć, jak zwierzęta reagują na zmieniające się warunki w trakcie pościgu, uwzględniając zarówno strategię ucieczki, jak i gonienia.

W robotyce i sztucznej inteligencji krzywe pościgowe mają ogromne znaczenie w kontekście planowania trajektorii robotów mobilnych oraz dronów. Drony autonomiczne, które muszą śledzić poruszające się cele, korzystają z algorytmów bazujących na tych krzywych, aby precyzyjnie dostosowywać swoje trajektorie do zmieniających się pozycji obiektów. Podobnie, w grach komputerowych krzywe pościgowe pozwalają na realistyczne modelowanie zachowania postaci NPC (Non-Player Characters), np. wrogów, którzy podążają za graczem.

W dziedzinie wojskowości krzywe pościgowe są fundamentalne dla projektowania systemów naprowadzania pocisków, takich jak pociski samonaprowadzające, które muszą precyzyjnie śledzić ruch celów. Ponadto, w systemach obrony przeciwrakietowej i przeciwlotniczej krzywe pościgowe są wykorzystywane do opracowywania trajektorii pocisków przechwytujących, które dążą do zniszczenia wroga.

W logistyce i transporcie krzywe pościgowe są stosowane do modelowania zachowań pojazdów, takich jak autonomiczne samochody czy statki morskie, które muszą ścigać lub unikać innych obiektów. Autonomiczne systemy nawigacji wykorzystują te krzywe do optymalizacji trajektorii pojazdów, tak aby zapewnić bezpieczeństwo i efektywność w ruchu drogowym czy na morzu. Z kolei w lotnictwie, w szczególności w strategiach ścigania celów w powietrzu, algorytmy krzywych pościgowych pomagają w projektowaniu systemów autopilotów i nawigacji. [[2]](#bibliografia)

W naukach przyrodniczych, takich jak fizyka płynów, krzywe pościgowe są używane do modelowania ruchu cząsteczek w przepływach, gdzie obiekty podążają za określonymi źródłami sił, na przykład w wirowaniu wody lub przepływach powietrza. Również w astrofizyce trajektorie ciał niebieskich, takich jak statki kosmiczne podążające za asteroidami, mogą być modelowane przy użyciu podobnych zasad matematycznych.

Z kolei w sportach drużynowych, takich jak piłka nożna, krzywe pościgowe pomagają w analizie strategii zawodników, którzy muszą podążać za przeciwnikiem, aby przejąć piłkę lub uniemożliwić mu strzał na bramkę. Tego typu modele mogą także wspierać trenerów w opracowywaniu strategii taktycznych. Zastosowania krzywych pościgowych są również szeroko obecne w analizie ruchu w animacjach komputerowych, gdzie symulowane postacie podążają za określonymi celami.

Dzięki swoim wszechstronnym zastosowaniom, krzywe pościgowe stanowią ważny element w rozwiązywaniu problemów optymalizacji w wielu dziedzinach.

oraz obiekt ścigający, którego wektor prędkości jest zawsze skierowany w stronę celu.
Będziemy chcieli znaleźć krzywą, po której będzie poruszał się obiekt ścigający.

Niech $\mathbf{a}$ będzie wektorem położenia obiektu uciekającego (czerwony kwadrat), natomiast $\mathbf{}b$ wektorem położenia obiektu ścigającego (niebieski trójkąt). Widzimy, że wektor prędkości $\mathbf{b'}$ jest równoległy do wektora $\mathbf{a-b}$. Stąd, normalizując te wektory, otrzymujemy:

$$
\frac{\mathbf{a} - \mathbf{b}}{|\mathbf{a} - \mathbf{b}|} \cdot \frac{\mathbf{b'}}{|\mathbf{b'}|} = 1
$$
Teraz weźmy $\mathbf{a}(t) = (u(t), v(t))$ (są to dane, zdefiniowane od początku funkcje) oraz $\mathbf{b}(t) = (x(t), y(t))$. Stąd otrzymujemy:

$$
\frac{(u(t) - x(t), v(t) - y(t))}{|\mathbf{a}(t) - \mathbf{b}(t)|} \cdot \frac{(x'(t), y'(t))}{|\mathbf{b'}(t)|} = 1
$$


Dla uproszczenia możemy przyjąć, że ścigający porusza się ze stałą jednostkową prędkością ($|b'|=1$). Dostajemy:
$$
\begin{equation}
\left\{
\begin{aligned}
&\frac{(u(t) - x(t)) x'(t) + (v(t) - y(t)) y'(t)}{\sqrt{(u(t) - x(t))^2 + (v(t) - y(t))^2}} = 1 \\
&(x'(t))^2 + (y'(t))^2 = 1
\end{aligned}
\right.
\end{equation}
$$

# Rozwiązanie analityczne - przykład 
Znajdziemy w sposób analityczny kształt krzywej, po której będzie poruszał się ścigający $\mathbf{g}$, jeśli uciekający $\mathbf{u}$ porusza się
po linii prostej (tj. $(x_u(t), y_u(t)) = (0,v_u\cdot t)$) , przyjmując prędkości uciekającego i goniącego za równe ($v_u=v_g=1$). Mamy zatem
$$ u(t) = (x_u(t), y_u(t))=(0,t) \text{ oraz } g(t) = (x_g(t), y_g(t))$$

Dla przejrzystości przyjmijmy oznaczenia $x_g(t)\sim x$, $y_g(t)\sim y$, $x'_g(t)\sim x'$ oraz $y'_g(t)\sim y'$. Wówczas otrzymujemy:
$$
\begin{align*}
&\frac{(-x, t - y)}{\sqrt{x^2 + (t - y)^2}} \cdot \frac{(x', y')}{1} = 1 \\
&-xx' + (t - y)y' = \sqrt{x^2 + (t - y)^2}\\
&x^2(x')^2-2xx'(t-y)y'+(t-y)^2(y')^2=x^2+(t-y)^2\\
&x^2((x')^2-1)+(t-y)^2((y')^2-1)-2x(t-y)x'y'=0\\
\end{align*}
$$
Wiedząc, że $((x')^2-1)=(x')^2-(x')^2-(y')^2=-(y')^2$ oraz $((y')^2-1)=(y')^2-(y')^2-(x')^2=-(x')^2$ dostajemy 
$$
\begin{align*}
&-x^2(y')^2-2x(t-y)x'y'-(t-y)^2(x')^2=0\\
&x^2(y')^2+2x(t-y)x'y'+(t-y)^2(x')^2=0\\
&(xy'+(t+y)x')^2=0\\
&xy'+(t+y)x'=0\\
&x\frac{dy(t)}{dt}+(t-y)\frac{dx(t)}{dt}=0
\end{align*}
$$
Mnożąc obustronnie przez $\frac{dx(t)}{dt}$ dostajemy:
$$xy'+(t-y)=0 \text{, gdzie } y'=\frac{\frac{dy}{dt}}{\frac{dx}{dt}}=\frac{dy}{dx}$$
Następnie pozbywamy się zmiennej t, korzystając ze wzoru na długość krzywej $ \int \sqrt{1+(y')^2}\,dx=s=vt=t$ 
$$xy'+\int \sqrt{1+(y')^2}\,dx-y=0$$
Różniczkujemy obustronnie i otrzymujemy:
$$
\begin{align*}
&y'+xy''+\sqrt{1+(y')^2}-y'=0\\
&xy''+\sqrt{1+(y')^2}=0
\end{align*}
$$
W naszym przykładzie możemy przyjąć, że $x>0$. Możemy zatem podzielić równanie przez $x$
$$
\begin{align*}
&y''+\frac{\sqrt{1+(y')^2}}{x}=0\\
&y''=-\frac{\sqrt{1+(y')^2}}{x}
\end{align*}$$
Otrzymane równanie różniczkowe drugiego rzędu rozwiążemy stosując podstawienie $v=y'$
$$
\begin{align*}
&v'=-\frac{\sqrt{1+v^2}}{x}\\
&\frac{dv}{dx}=-\frac{\sqrt{1+v^2}}{x}\\
&\frac{dv}{\sqrt{1+v^2}}=-\frac{1}{x}dx\\
&\sinh^{-1}(v)=-\ln(x)+C\\
&v=\sinh(-\ln(x)+C)\\
&v=\frac{1}{2}(e^{(C-\ln(x))}-(e^{-(C-\ln(x))})\\
&v=\frac{1}{2}(\frac{e^C}{x}-xe^{-C})=y'\\
&y'=\frac{dy}{dx}=\frac{1}{2}(\frac{e^C}{x}-xe^{-C})\\
&y=\frac{1}{2}(e^C\ln(x)-\frac{1}{2}e^{-C}x^2)+E
\end{align*}
$$
Możemy dodatkowo uporządkować stałe w równaniu. Widzimy, że $y=A\ln(x)-Bx^2+E$, gdzie 
$A=\frac{e^C}{2}$ oraz $B=\frac{e^{-C}}{4}=\frac{1}{4e^C}=\frac{1}{8A}$. Przyjmujemy również założenia początkowe $y(x_0)=y_0$ oraz $y'(x_0)=\frac{y_0}{x_0}$. Możemy zatem zapisać:
$$
\begin{equation}
\left\{
\begin{aligned}
&y=C_1+C_2x^2-\frac{\ln(x)}{8C_2}\\
&y(x_0)=y_0\\
&y'(x_0)=\frac{y_0}{x_0}
\end{aligned}
\right.
\end{equation}
$$
Rozwiązaniem powyższego zagadnienia jest funkcja [[3]](#bibliografia)
$$
y=\frac{1}{4}\left[\left(y_0+\sqrt{x_0^2+y_0^2}\right)\left(\frac{x}{x_0}\right)^2+\left(y_0-\sqrt{x_0^2+y_0^2}\right)\ln\left(\left(\frac{x}{x_0}\right)^2\right)+3y_0-\sqrt{x_0^2+y_0^2}\right]
$$
Narysujemy teraz przykładową krzywą dla położenia początkowego goniącego $x_0=y_0=3$. Wiemy, że będzie się on poruszał po krzywej danej wzorem:
$$
y=\frac{1}{4}\left[\left(3+\sqrt{18}\right)\left(\frac{x^2}{9}\right)+\left(3-\sqrt{18}\right)\ln\left(\frac{x^2}{9}\right)+9-\sqrt{18}\right]
$$

```python
import matplotlib.pyplot as plt
import numpy as np


# Tworzenie wykresu prostego układu współrzędnych
plt.figure(figsize=(8, 6))

# Rysowanie osi X i Y
plt.axhline(0, color='black',linewidth=1)  # Oś X
plt.axvline(0, color='black',linewidth=1)  # Oś Y

# Ustawienie zakresu
plt.xlim(-1, 10)
plt.ylim(0, 10)

# Dodanie etykiet
plt.xlabel('x')
plt.ylabel('y')
plt.title('')

# Siatka
plt.grid(True)

#punkty 
x_apoint = 0
y_apoint = 0
plt.scatter(x_apoint, y_apoint, color='blue', s=200, label=f'Punkt startowy uciekającego')

x_bpoint = 3
y_bpoint = 3
plt.scatter(x_bpoint, y_bpoint, color='red', s=200, label=f'Punkt startowy goniącego')

plt.axvline(0, color='purple', linestyle='--', linewidth=2, label='trasa uciekającego') 

# Dodanie funkcji
x_values = np.linspace(0.001, 3, 500)  # Przedział wartości x (od 0.01 do 3)
sqrt_3_squared = np.sqrt(3**2 + 3**2)  # sqrt(3^2 + 3^2)
f_values = (1/4) * (
    (3 + sqrt_3_squared) * (x_values / 3)**2 +
    (3 - sqrt_3_squared) * np.log((x_values / 3)**2) +
    3 * 3 - sqrt_3_squared
)
plt.title('Krzywa pościgowa - rozwiazanie analityczne')
plt.plot(x_values, f_values, label=r"trasa goniącego", color='orange')  # Rysowanie
plt.legend()
```

Otrzymana trasa ruchu obiektu goniącego jest zgodna z prognozami. 

# Podejście numeryczne - metoda Eulera

Spróbujemy teraz rozważyć podejście numeryczne w szukaniu krzywych pościgowych. Zaczniemy od wyjaśnienia działania algorytmu do poszukiwania krzywych metodą numeryczną. Na początek zajmiemy się przybliżaniem położenia obiektu goniącego przy pomocy metody Eulera.

Do poprawnego zaprogramowania algorytmu niezbędny będzie wyprowadzony wcześniej układ równań, który opisuje położenie goniącego w czasie
$$
\begin{equation}
\left\{
\begin{aligned}
&\frac{(u(t) - x(t)) x'(t) + (v(t) - y(t)) y'(t)}{\sqrt{(u(t) - x(t))^2 + (v(t) - y(t))^2}} = 1 \\
&(x'(t))^2 + (y'(t))^2 = 1
\end{aligned}
\right.
\end{equation}
$$
Stosując metodę Eulera możemy przybliżyć pochodne składowych położenia. Dyskretyzujemy czas, a odstęp między kolejnymi momentami w czasie oznaczamy jako $\Delta t$. Oznaczmy przez $t_0$ czas teraźniejszy, natomiast przez $t_1$ czas po upłynięciu $\Delta t$. Wtedy:
$$
\begin{equation}
\begin{aligned}
x'(t)~=\frac{x(t_1)-x(t_0)}{\Delta t}\\
y'(t)~=\frac{y(t_1)-y(t_0)}{\Delta t}
\end{aligned}
\end{equation}
$$

Oznaczając $u(t_0)\sim u_0$, $u(t_1)\sim u_1$, $v(t_0)\sim v_0$, $x(t_0)\sim x_0$, $x(t_1)\sim x_1$,
$y(t_0)\sim y_0$, $y(t_1)\sim y_1$ oraz podstawiając do układu równań otrzymujemy:
$$
\begin{equation}
\left\{
\begin{aligned}
&\frac{(u_1 - x_1) \frac{x_1-x_0}{\Delta t} + (v_1 - y_1) \frac{y_1-y_0}{\Delta t}}{\sqrt{(u_1 - x_1)^2 + (v_1 - y_1)^2}}-1 = 0 \\
&\left(\frac{x_1-x_0}{\Delta t}\right)^2 + \left(\frac{y_1-y_0}{\Delta t}\right)^2-1 = 0
\end{aligned}
\right.
\end{equation}
$$


Mając odpowiednie przybliżenie pochodnych oraz układ równań możemy zastosować algorytm:

1. Ustawiamy $t_0$, a także $v_1, y_1, x_0, y_0$, które są ustalone.

2. Podstawiamy je do naszych równań i za pomocą funkcji Pythona (w naszym przypadku fsolve)       znajdujemy $x_1$ oraz $y_1$.
     
3. Stawiamy $x_0=x_1, y_0=y_1, v_1=v_2, u_1=u_2, t_2=t_1+\Delta t$ i powtarzamy algorytm.

Zadane punkty położenia uciekającego a także obliczone za pomocą algorytmu punkty położenia goniącego w danych momentach czasu przechowujemy w tablicach i za pomocą tych punktów rysujemy trasy obiektów.

Zaimplementujemy teraz nasz algorytm w kodzie Python do znalezienia krzywej pościgowej z przykładu, który rozwiązaliśmy analitycznie. Spodziewamy się, że otrzymana krzywa powinna wyglądać podobnie do tej otrzymanej w sposób analityczny - będzie to wskazywało na skuteczność metody numerycznej.

```python
from math import sqrt
import matplotlib.pyplot as plt
import numpy as np
import warnings

warnings.filterwarnings("ignore", category=RuntimeWarning)

from scipy.optimize import fsolve
def przyblizanie1(xt0, yt0,ut1,vt1,dt): #funkcja ktora pomaga nam znalezc rozwiazania numeryczne x1 i y1 
    # Funkcja definiująca układ równań
    def equations(vars, xt0, yt0,ut1,vt1,dt):
        xt1, yt1 = vars
        eq1 = ((ut1-xt1)*((xt1-xt0)/dt)+(vt1-yt1)*((yt1-yt0)/dt))/(sqrt((ut1-xt1)**2+(vt1-yt1)**2))-1 #rownanie 1
        eq2 = ((xt1-xt0)/dt)**2+((yt1-yt0)/dt)**2-1 #rownanie 2 
        return [eq1, eq2]
    # Przybliżenie początkowe
    initial_guess = [xt0+1/10000,yt0+1/10000]

    # Rozwiązywanie układu równań
    solution = fsolve(equations, initial_guess, args=(xt0, yt0,ut1,vt1,dt))

    return solution

#tablica wartosci u i v w dla poszczegolnych t (przypadek dla (u,v)=(0,t))
dt=1/10000 #ustawiamy krok 
tabv=[] #tablica wartosci v dla (uzupelnimy ja wartosciami v dla kolejnych krokow czasu)
v0=0 #polozenie poczatkowe (v=0)
tabv.append(v0)
for i in range (1, 100000): # uzupelniamy polozenie czastki uciekajacej po kolejnych krokach
    v0=v0+dt
    tabv.append(v0)

def euler5():
    rozwiazanie=[]
    vt1=tabv[1] #polozenie uciekiniera
    ut1=0 #polozenie uciekiniera, ale u w tym wypadku jest stale rowne 0
    xt0=3 #polozenie poczatkowe goniacego
    yt0=3 #polozenie poczatkowe goniacego
    for i in range (0,99999):
        para=przyblizanie1(xt0, yt0,ut1,vt1,dt)
        rozwiazanie.append(para[0]) #tablica rozwiazanie jest postaci x1,y1,x2,y2,x3,y3,...
        rozwiazanie.append(para[1])
        xt0=para[0] #stawiamy nowe wartosci x0 i y0 do kolejnej iteracji
        yt0=para[1]
        ut1=0 # w tym przypadku nadal stale rowne 0
        vt1=tabv[i+1] #bierzemy v uciekajacego po nastepnym kroku
    return(rozwiazanie) #zwracamy tablice rozwiazan w postaci x1,y1,x2,y2,x3,y3

wynik=euler5()


x = wynik[::2]  # Co drugi element zaczynając od 0 to x
y = wynik[1::2] # Co drugi element zaczynając od 1 to y

# Tworzenie wykresu
plt.figure(figsize=(8, 6))
plt.plot(x, y, color='blue', label='trasa goniącego - z metody numerycznej')
plt.xlim(-1, 10)  # Zakres osi X
plt.ylim(-1, 10)  # Zakres osi Y

# Dodanie funkcji z rozwiazania analitycznego
x_values = np.linspace(0.001, 3, 15)  # Przedział wartości x (od 0.01 do 3)
sqrt_3_squared = np.sqrt(3**2 + 3**2)  # sqrt(3^2 + 3^2)
f_values = (1/4) * (
    (3 + sqrt_3_squared) * (x_values / 3)**2 +
    (3 - sqrt_3_squared) * np.log((x_values / 3)**2) +
    3 * 3 - sqrt_3_squared
)

plt.scatter(x_values, f_values, label=r"trasa goniącego - z metody analitycznej", color='orange')  # Rysowanie

x_values = np.logspace(np.log10(0.001), np.log10(0.15), 9)
sqrt_3_squared = np.sqrt(3**2 + 3**2)  # sqrt(3^2 + 3^2)
f_values = (1/4) * (
    (3 + sqrt_3_squared) * (x_values / 3)**2 +
    (3 - sqrt_3_squared) * np.log((x_values / 3)**2) +
    3 * 3 - sqrt_3_squared
)
plt.scatter(x_values, f_values, color='orange')  # Rysowanie

# Dodanie etykiet i tytułu
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Krzywa pościgowa - metoda Eulera')

# Pokazanie wykresu
plt.legend()
plt.grid(True)
plt.show()
```

# Rozwiązanie numeryczne - metoda centralna

Inną metodą, która pozwoli nam na numeryczne wyznaczenie trasy ścigającego jest metoda centralna. Spróbujemy za jej pomocą wyznaczyć tę trasę w tej samej sytuacji co w poprzednim przykładzie. Nasz program będzie wyglądał bardzo podobnie, natomiast do przybliżania pochodnej zastosujemy właśnie tę metodę.

Metoda centralna do przybliżania pochodnej polega na obliczeniu różnicy funkcji w dwóch punktach symetrycznych (w kroku poprzedzającym i następnym) względem punktu, w którym chcemy obliczyć pochodną. Wzór na przybliżenie pochodnej przy pomocy metody centralnej wygląda następująco:
$$f'(x) \approx \frac{f(x+\Delta t) - f(x-\Delta t)}{2\Delta t}$$
, gdzie $f(x)$ to funkcja, której pochodną obliczamy.

Metoda centralna powinna być dokładniejsza niż metoda Eulera, ponieważ wykorzystuje informację z obu stron punktu co zmniejsza błąd przybliżenia.







```python
from math import sqrt
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import fsolve

#((ut1-xt1)*((x1-x0)/dt)+(vt1-yt1)*((yt1-yt0)/dt))/(sqrt((u1t-x1t)**2+(v1t-y1t)**2))-1=0
#((xt1-xt0)/dt)**2+((yt1-yt0)/dt)**2-1=0


def przyblizanie2(xt0, yt0, xt1, yt1, ut1, vt1, dt):
    # Funkcja definiująca układ równań
    def equations(vars, xt0, yt0, xt1, yt1, ut1, vt1, dt):
        xt2, yt2 = vars
        eq1 = ((ut1-xt1)*((xt2-xt0)/(2*dt))+(vt1-yt1)*((yt2-yt0)/(2*dt)))/(sqrt((ut1-xt1)**2+(vt1-yt1)**2))-1
        eq2 = ((xt2-xt0)/(2*dt))**2+((yt2-yt0)/(2*dt))**2-1
        return [eq1, eq2]
    # Przybliżenie początkowe
    initial_guess = [xt1+1/1000,yt1+1/1000]

    # Rozwiązywanie układu równań
    solution = fsolve(equations, initial_guess, args=(xt0, yt0, xt1, yt1, ut1, vt1, dt),xtol=1e-6)

    return solution

#tablica wartosci u i v w dla poszczegolnych t (przypadek dla (u,v)=(0,t))
dt=1/10000 
tabv=[]
v0=0
tabv.append(v0)
for i in range (1, 100000):
    v0=v0+dt
    tabv.append(v0)
   # print(-i)

def euler_central5():
    rozwiazanie = []
    vt1 = tabv[0] 
    ut1 = 0
    xt0, yt0 = 3, 3 
    xt1, yt1 = 3 - 1 / 20000, 3 + 1 / 20000 
    rozwiazanie.append(xt0)
    rozwiazanie.append(yt0)
    rozwiazanie.append(xt1)
    rozwiazanie.append(yt1)



    for i in range(99999): 
        para = przyblizanie2(xt0, yt0, xt1, yt1, ut1, vt1, dt)

     
        rozwiazanie.append(para[0])
        rozwiazanie.append(para[1])

        # Aktualizacja do nastepnej iteracji
        xt0, yt0 = xt1, yt1
        ut1, vt1 = 0, tabv[i + 1]  
        xt1, yt1 = para[0], para[1] 


    return rozwiazanie


wynik=euler_central5()


if wynik: 
    x = wynik[::2]  
    y = wynik[1::2] 
   
    plt.figure(figsize=(8, 6))
    
    # Dodanie funkcji z rozw analitycznego
    x_values = np.linspace(0.001, 3, 15)  # Przedział wartości x (od 0.01 do 3)
    sqrt_3_squared = np.sqrt(3**2 + 3**2)  # sqrt(3^2 + 3^2)
    f_values = (1/4) * (
        (3 + sqrt_3_squared) * (x_values / 3)**2 +
        (3 - sqrt_3_squared) * np.log((x_values / 3)**2) +
        3 * 3 - sqrt_3_squared
    )

    plt.scatter(x_values, f_values, label=r"trasa goniącego - z metody analitycznej", color='orange')  # Rysowanie

    x_values = np.logspace(np.log10(0.001), np.log10(0.15), 9)
    sqrt_3_squared = np.sqrt(3**2 + 3**2)  # sqrt(3^2 + 3^2)
    f_values = (1/4) * (
        (3 + sqrt_3_squared) * (x_values / 3)**2 +
        (3 - sqrt_3_squared) * np.log((x_values / 3)**2) +
        3 * 3 - sqrt_3_squared
    )
    plt.scatter(x_values, f_values, color='orange')  # Rysowanie



    plt.plot(x, y, color='blue', label='trasa wyznaczona z metody centralnej')
    plt.xlim(-1, 10)
    plt.ylim(-1, 10)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Krzywa pościgowa - metoda centralna')
    plt.legend()
    plt.grid(True)
    plt.show()
    print()
else:
    print("Brak danych")
```

# Metoda Eulera - błąd i zbieżność 

Metoda Eulera, jako metoda pierwszego rzędu, charakteryzuje się błędem lokalnym rzędu $O(h^2)$, co oznacza, że błąd na każdym kroku zależy kwadratowo od wielkości kroku $h$. Jednak błąd globalny, wynikający z sumowania błędów lokalnych na przestrzeni czasu, jest rzędu $O(h)$, co oznacza, że maleje on liniowo wraz ze zmniejszaniem kroku. Niestabilności w metodzie mogą wynikać ze zbyt dużego kroku, co zwiększa akumulację błędów lub też silnej zależności wyników od warunków początkowych i charakterystyki równań. [[4]](#bibliografia)


Na ewentualną kumulacje błędu wpływ może mieć rownież funkcja fsolve [[5]](#bibliografia), używana w naszym algorytmie do rozwiązania układu równań róźniczkowych. Błąd funkcji fsolve może wynikać z faktu, iż w wewnętrznych obliczeniach funkcja używa numerycznie aproksymowanej pochodnej funkcji (metoda Newtona-Raphsona [[6]](#bibliografia)). Jeśli funkcja w określonym zakresie jest bardzo płaska lub horyzontalna, algorytm może zwrócić błędny wynik albo nie zbiegać do rozwiązania.

Dobór odpowiedniego punktu początkowego (ang. initial guess) ma kluczowe znaczenie dla zbieżności algorytmu fsolve. Algorytmy Newtona-Raphsona są znane ze swojej lokalnej zbieżności, czyli zbieżności tylko wtedy, gdy punkt początkowy znajduje się wystarczająco blisko rozwiązania układu równań. Zbyt daleki punkt początkowy może prowadzić do niepoprawnych wyników lub nawet do braku zbieżności algorytmu, zwłaszcza w przypadku nieliniowych równań, które mogą mieć wiele rozwiązań.
Z tego powodu, aby zapewnić skuteczność i stabilność algorytmu, należy starać się dostarczyć fsolve punkt początkowy, który znajduje się blisko oczekiwanego rozwiązania. W przeciwnym przypadku, w sytuacjach, gdzie dostępne są jedynie ogólne informacje o funkcji, może być konieczne przeprowadzenie kilku prób z różnymi punktami początkowymi w celu znalezienia najbardziej odpowiedniego. Dodatkowo, w przypadku układów o wielu rozwiązaniach, dobór początkowego punktu może wpływać na to, które rozwiązanie zostanie znalezione przez algorytm.

Zastosowanie metody Eulera w połączeniu z fsolve powinno być zbieżne, należy jednak zadbać o odpowiednia liczbę iteracji, dokładność implementacji równań różniczkowych oraz odpowiednie dobranie warunków funkcji fsolve.


Metody numeryczne są powszechnie stosowane do rozwiązywania równań różniczkowych, szczególnie w przypadkach, gdy rozwiązania analityczne są trudne lub niemożliwe do uzyskania. Kluczowym aspektem analizy tych metod jest ocena ich dokładności, czyli badanie, jak dobrze rozwiązanie numeryczne przybliża rzeczywiste rozwiązanie problemu w zależności od wybranych parametrów, takich jak krok czasowy $\Delta t$. Jednym z istotnych narzędzi w tej analizie jest badanie maksymalnego błędu
$E_{max}$ , czyli największej różnicy między rozwiązaniem numerycznym a rozwiązaniem analitycznym w danym przedziale czasowym.

Dla wielu metod numerycznych, takich jak metoda Eulera, maksymalny błąd  $E_{max}$ zależy od liczby podziałów czasu $N_t$ (która zależy od kroku czasowego $\Delta t$) zgodnie z równaniem:
$$
E_{max} \sim C (N_t)^{-p},
$$
gdzie $C$ to stała zależna od problemu, a $p$ to rząd metody numerycznej, który określa szybkość zbieżności metody.

Aby zbadać zależność $E_{max}$ od $\Delta t$, często stosuje się wykres w skali log-log. Wykres ten pozwala zinterpretować zależność jako prostą linię (w naszym przypadku będzie to jej przybliżenie), gdzie nachylenie tej linii odpowiada rzędowi metody $p$. Matematycznie, biorąc logarytm obu stron równania, otrzymujemy:
$$
\log(E_{max}) = \log(C) - p \cdot \log(N_t).
$$

Wykres $ \log(E_{max}) $ względem $ \log(N_t)$ powinien być liniowy, a współczynnik kierunkowy tej linii wynosi $-p$, co umożliwia oszacowanie rzędowości metody.

Analiza błędu maksymalnego na wykresie log-log jest nie tylko narzędziem do oceny jakości metody numerycznej, ale również pozwala na porównanie różnych metod i ich efektywności. Metody wyższego rzędu (większe $p$) zazwyczaj oferują większą dokładność przy tej samej wartości kroku $\Delta t$, ale mogą być bardziej kosztowne obliczeniowo. Z tego względu dobór odpowiedniej metody numerycznej zależy od kompromisu między dokładnością a wydajnością.

Narysujemy teraz wykres $E_{max}$ w zależności od liczby podziałów $N$ (która zależy od kroku czasowego $\Delta t$. Następnie wykres przedstawimy w skali log-log i wyznaczymy $p$. Spodziewamy się, że $p$ będzie w naszym przypadku mniejsze od 1, gdyż w naszym problemie mamy układ równań, który musimy rozwiązywać numerycznie za pomocą fsolve, a to może mieć wpływ na błąd metody.


```python
from math import sqrt
import matplotlib.pyplot as plt
import numpy as np
import warnings
import math

warnings.filterwarnings("ignore", category=RuntimeWarning)

from scipy.optimize import fsolve
def przyblizanie(xt0, yt0,ut1,vt1,dt): #funkcja ktora pomaga nam znalezc rozwiazania numeryczne x1 i y1 
    # Funkcja definiująca układ równań
    def equations(vars, xt0, yt0,ut1,vt1,dt):
        xt1, yt1 = vars
        eq1 = ((ut1-xt1)*((xt1-xt0)/dt)+(vt1-yt1)*((yt1-yt0)/dt))/(sqrt((ut1-xt1)**2+(vt1-yt1)**2))-1 #rownanie 1
        eq2 = ((xt1-xt0)/dt)**2+((yt1-yt0)/dt)**2-1 #rownanie 2 
        return [eq1, eq2]
    # Przybliżenie początkowe
    initial_guess = [xt0+1/10000,yt0+1/10000]

    # Rozwiązywanie układu równań
    solution = fsolve(equations, initial_guess, args=(xt0, yt0,ut1,vt1,dt), xtol=1e-6) #xtol=1e-6 ZMNIEJSZA DOKLADNOSC

    return solution

#tablica wartosci u i v w dla poszczegolnych t (przypadek dla (u,v)=(0,t))
dt=1/10 #ustawiamy krok 

#czas=10 sekund dt*X=czas => X=czas\dt

def euler(dt):

    tabv=[] #tablica wartosci v dla (uzupelnimy ja wartosciami v dla kolejnych krokow czasu)
    v0=0 #polozenie poczatkowe (v=0)
    tabv.append(v0)
    for i in range (1, int(10/dt)): # uzupelniamy polozenie czastki uciekajacej po kolejnych krokach
        v0=v0+dt
        tabv.append(v0)
    rozwiazanie=[]
    vt1=tabv[1] #polozenie uciekiniera
    ut1=0 #polozenie uciekiniera, ale u w tym wypadku jest stale rowne 0
    xt0=3 #polozenie poczatkowe goniacego
    yt0=3 #polozenie poczatkowe goniacego
    for i in range (0,int(10/dt)-1):
        para=przyblizanie(xt0, yt0,ut1,vt1,dt)
        rozwiazanie.append(para[0]) #tablica rozwiazanie jest postaci x1,y1,x2,y2,x3,y3,...
        rozwiazanie.append(para[1])
        xt0=para[0] #stawiamy nowe wartosci x0 i y0 do kolejnej iteracji
        yt0=para[1]
        ut1=0 # w tym przypadku nadal stale rowne 0
        vt1=tabv[i+1] #bierzemy v uciekajacego po nastepnym kroku
    return(rozwiazanie) #zwracamy tablice rozwiazan w postaci x1,y1,x2,y2,x3,y3

#wynik=euler(1/100)


#x = wynik[::2]  # Co drugi element zaczynając od 0 to x
#y = wynik[1::2] # Co drugi element zaczynając od 1 to y



#y_prawdziwe = (1/4)*((3+math.sqrt(18))*((x_e[i]**2)/9) + (3-math.sqrt(18))*math.log((x_e[i]**2)/9) + 9 - math.sqrt(18))

def maximum(dt):
    wynik=euler(dt)
    x = wynik[::2]  # Co drugi element zaczynając od 0 to x
    y = wynik[1::2] # Co drugi element zaczynając od 1 to y
    bledy=[]
    for i in range (0, len(x)):
        blad=y[i]- (1/4)*((3+math.sqrt(18))*((x[i]**2)/9) + (3-math.sqrt(18))*math.log((x[i]**2)/9) + 9 - math.sqrt(18))
        if(x[i]>0.05):
            bledy.append(abs(blad))
    return max(bledy)

dziedzina=[]
bledy=[]
for i in range(100,1000,100):
    dziedzina.append(10/(1/i))
    bledy.append(maximum(1/i))

#dziedzina=[10/(1/100),10/(2/100),10/(3/100),10/(4/100)]
#bledy=[maximum(1/100),maximum(2/100),maximum(3/100),maximum(4/100) ]


plt.figure(figsize=(8, 6))
plt.plot(dziedzina, bledy, color='blue', label='bład')
# Dodanie etykiet i tytułu
plt.xlabel('N')
plt.ylabel('Emax')
plt.title('Wykres błędów maksymalnych w zależności od liczby podziałów')

# Pokazanie wykresu
plt.legend()
plt.grid(True)
plt.show()

# Tworzenie wykresu
plt.figure(figsize=(8, 6))
plt.loglog(dziedzina, bledy, color='red', label='bład')
#plt.xlim(-1, 10)  # Zakres osi X
#plt.ylim(-1, 10)  # Zakres osi Y

# Dodanie etykiet i tytułu
plt.xlabel('N')
plt.ylabel('Emax')
plt.title('Wykres błedów w skali log-log')

# Pokazanie wykresu
plt.legend()
plt.grid(True)
plt.show()


# Wyznaczanie nachylenia prostej (rzędowość metody)
coeffs = np.polyfit(np.log(dziedzina), np.log(bledy), 1)  # Dopasowanie prostej
slope = (-1)*coeffs[0]
print(f"Rzędowość metody: {slope:.2f}")
```

# Pościg, gdy uciekający porusza się po okręgu.

Załóżmy teraz, że uciekający porusza się po okręgu, a ścigający zaczyna swój ruch poza tym okręgiem.
W tym przypadku znalezienie trasy goniącego byłoby zdecydowanie trudniejsze niż w poprzednim przykładzie, jednakże w metodzie numerycznej wystarczy jedynie zmienić trasę uciekającego, a cały algorytm powinien zadziałać poprawnie. Intuicja wskazywałaby, że ścigający zacznie zbliżać się do okręgu, a gdy w pewnym punkcie dogoni uciekającego, to będzie poruszał się po okręgu razem z nim.

```python
from math import sqrt, cos, sin
import matplotlib.pyplot as plt
import numpy as np
import warnings

warnings.filterwarnings("ignore", category=RuntimeWarning)


from scipy.optimize import fsolve
def przyblizanie(xt0, yt0,ut1,vt1,dt):
    # Funkcja definiująca układ równań
    def equations(vars, xt0, yt0,ut1,vt1,dt):
        xt1, yt1 = vars
        eq1 = ((ut1-xt1)*((xt1-xt0)/dt)+(vt1-yt1)*((yt1-yt0)/dt))/(sqrt((ut1-xt1)**2+(vt1-yt1)**2))-1
        eq2 = ((xt1-xt0)/dt)**2+((yt1-yt0)/dt)**2-1
        return [eq1, eq2]
    # Przybliżenie początkowe
    initial_guess = [xt0+1/10000,yt0+1/10000]

    # Rozwiązywanie układu równań
    solution = fsolve(equations, initial_guess, args=(xt0, yt0,ut1,vt1,dt))

    return solution

dt=1/1000
u0=1
tabv=[]
tabu=[]
v0=0
t0=0
tabv.append(v0)
tabu.append(u0)
for i in range (1, 100000): #teraz punkt uciekajacy porusza sie po okregu zaczynajac od punktu (1,0)
    v0=(cos(t0))
    tabv.append(v0)
    u0=(sin(t0))
    tabu.append(u0)
    t0=t0+dt


def euler():
    rozwiazanie=[]
    vt1=tabv[1]
    ut1=tabu[1]
    xt0=2 #polozenie poczatkowe goniacego
    yt0=2 #polozenie poczatkowe goniacego
    for i in range (0,99999):
        para=przyblizanie(xt0, yt0,ut1,vt1,dt)
        rozwiazanie.append(para[0])
        rozwiazanie.append(para[1])
        xt0=para[0]
        yt0=para[1]
        ut1=tabu[i+1]
        vt1=tabv[i+1]
    return(rozwiazanie) #zwracamy tablice rozwiazan w postaci x1,y1,x2,y2,x3,y3

wynik=euler()


x = wynik[::2]  # Co drugi element zaczynając od 0 to x
y = wynik[1::2] # Co drugi element zaczynając od 1 to y

# Tworzenie wykresu
plt.figure(figsize=(8, 6))
plt.plot(x, y, color='blue', label='trasa goniacego')
plt.xlim(-3, 3)  # Zakres osi X 
plt.ylim(-3, 3)  # Zakres osi Y

# Dodanie etykiet i tytułu
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Metoda Eulera - przykład z okręgiem')

# Liczba punktów
num_points = 20

# Parametry okręgu
center = (0, 0)
radius = 1

# Generowanie kątów
angles = np.linspace(0, 2 * np.pi, num_points, endpoint=False)

# Obliczanie współrzędnych punktów
x_points = center[0] + radius * np.cos(angles)
y_points = center[1] + radius * np.sin(angles)

plt.scatter(x_points, y_points, color='orange', label='trasa uciekajacego') 

# Pokazanie wykresu
plt.legend()
plt.grid(True)
plt.show()
```

Intuicyjne rozważania wydają się być poprawne. Goniący zbliża się do uciekającego, a im jest bliżej, tym jego trasa coraz bardziej przypomina okrag.

# Więcej ścigających - udoskonalenie algorytmu
Drobne poprawki w algorytmie, czyli m.in. wstawienie trasy obiektu ściganego (tablicy zawierającej współrzędne położenia w poszczególnych czasach) jako parametr do funkcji wyznaczającej trasę goniącego sprawiają, że teraz jesteśmy w stanie w prosty sposób symulować i obrazować sytuacje, gdy mamy więcej niż jednego goniącego i/lub więcej niż jednego uciekającego.

# Przykład - gonitwa myszy, kota i psa.
Wyobraźmy sobię sytuację, w której mysz biegnie do swojej dziury, poruszając się slalomem (droga sinusoidalna). W chwili, gdy mysz jest już zaledwie 10 sekund od swojej dziury (położenie początkowe w symulacji (0,0)) rusza za nią kot (z położenia początkowego (2,-1)), a za kotem, z położenia startowego (-2, 0) rusza pies. Kot goni mysz, a pies goni kota. Naszym celem jest wyznaczenie na wykresie trasy kota i psa oraz sprawdzenie, czy przez następne 10 sekund (czyli aż mysz nie wbiegnie do dziury) kot złapie mysz lub pies złapie kota. W przypadku gdy nikt nikogo nie złapie sprawdzimy, czy po 10 sekundach kot był bliżej myszy, czy pies bliżej kota. Zakładamy, że pies i kot poruszają się z tą samą, jednostkową prędkością.

Możemy przypuszczać, że trasa zarówno kota, jak i psa po pewnym czasie zacznie się zachowywać podobnie do funkcji sinus (zacznie oscylować).

Wykorzystujemy nasz usprawniony algorytm do wyznaczenia tras psa i kota. Chcemy, by w tym wypadku obliczenia były bardzo dokładne, dlatego ustawimy bardzo mały krok $dt$, mimo że przez to obliczenia zajmą nam więcej czasu.

```python
from math import sqrt, cos, sin
import matplotlib.pyplot as plt
import numpy as np
import warnings

warnings.filterwarnings("ignore", category=RuntimeWarning)


from scipy.optimize import fsolve
def przyblizanie(xt0, yt0,ut1,vt1,dt):
    # Funkcja definiująca układ równań
    def equations(vars, xt0, yt0,ut1,vt1,dt):
        xt1, yt1 = vars
        eq1 = ((ut1-xt1)*((xt1-xt0)/dt)+(vt1-yt1)*((yt1-yt0)/dt))/(sqrt((ut1-xt1)**2+(vt1-yt1)**2))-1
        eq2 = ((xt1-xt0)/dt)**2+((yt1-yt0)/dt)**2-1
        return [eq1, eq2]
    # Przybliżenie początkowe
    initial_guess = [xt0+1/10000,yt0+1/10000]

    # Rozwiązywanie układu równań
    solution = fsolve(equations, initial_guess, args=(xt0, yt0,ut1,vt1,dt))

    return solution 
dt=1/10000 #[s], czas calkowity to 1/1000*10000 czyli 10s
u0=0
tabv=[]
tabu=[]
v0=0 #punkt startowy myszy
t0=0 #punkt startowy myszy
tabv.append(v0)
tabu.append(u0)
for i in range (1, 100000): #mysz ucieka slalomem (sinus)
    v0=(sin(t0))
    tabv.append(v0)
    u0=(t0)
    tabu.append(u0)
    t0=t0+dt


def euler2(tabu2,tabv2,a,b): # funkcja zalezna od trasy uciekajacego i punktu startowego goniacego
    rozwiazanie=[]
    vt1=tabv2[1]
    ut1=tabu2[1]
    xt0=a #polozenie poczatkowe goniacego
    yt0=b #polozenie poczatkowe goniacego
    for i in range (0,len(tabu2)-1):
        para=przyblizanie(xt0, yt0,ut1,vt1,dt)
        rozwiazanie.append(para[0])
        rozwiazanie.append(para[1])
        xt0=para[0]
        yt0=para[1]
        ut1=tabu2[i+1]
        vt1=tabv2[i+1]
    return(rozwiazanie) #zwracamy tablice rozwiazan w postaci x1,y1,x2,y2,x3,y3

wynik=euler2(tabu,tabv,2,-1) #trasa kota


x = wynik[::2]  # Co drugi element zaczynając od 0 to x
y = wynik[1::2] # Co drugi element zaczynając od 1 to y

wynik2=euler2(x,y,-2,-0) #trasa psa

x2 = wynik2[::2]  # Co drugi element zaczynając od 0 to x
y2 = wynik2[1::2] # Co drugi element zaczynając od 1 to y

# Tworzenie wykresu
plt.figure(figsize=(8, 6))
plt.plot(x, y, color='blue', label='trasa kota') #kot
plt.plot(x2, y2, color='red', label='trasa psa') #pies
plt.plot(tabu, tabv, color='green', label='trasa myszy') #mysz
plt.xlim(-3, 15)  # Zakres osi X 
plt.ylim(-3, 3)  # Zakres osi Y

# Dodanie etykiet i tytułu
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Przykład - mysz, kot i pies')

# Pokazanie wykresu
plt.legend()
plt.grid(True)
plt.show()

#kto po 10 sekundach bedzie blizej: kot myszy czy pies kota?

xmyszy=tabu[len(tabu)-1]
ymyszy=tabv[len(tabv)-1]

xkota=x[len(x)-1]
ykota=y[len(y)-1]

xpsa=x2[len(x2)-1]
ypsa=y2[len(y2)-1]

dkotmysz=sqrt((xmyszy-xkota)**2+(ymyszy-ykota)**2)
dpieskot=sqrt((xkota-xpsa)**2+(ykota-ypsa)**2)

#print(dkotmysz)
#print(dpieskot)
print(f'dpieskot < dkotmysz : {dpieskot<dkotmysz}') #sprawdzamy kto jest blizej

zlapano=False

for i in range(1,99999): #sprawdzimy czy w czasie poscigu pies zlapal kota lub kot mysz
    xmyszy=tabu[len(tabu)-1]
    ymyszy=tabv[len(tabv)-1]

    xkota=x[i-1]
    ykota=y[i-1]

    xpsa=x2[i-1]
    ypsa=y2[i-1]

    dkotmysz=sqrt((xmyszy-xkota)**2+(ymyszy-ykota)**2)
    dpieskot=sqrt((xkota-xpsa)**2+(ykota-ypsa)**2)
    if(dkotmysz<1/3 or dpieskot<1/3): #ustalmy zasieg psa i kota na 1/3
        zlapano=True

print(f'czy zlapano: {zlapano}')
```

Nasze przypuszczenia co do kształtu trasy zwierząt okazały się zgodne z prawda. Z obliczeń wynika natomiast, że kot nie dogoni myszy ani pies nie dogoni kota, natomiast pod koniec pogoni to pies będzie znajdował się bliżej kota niż kot myszy.

# Gonitwa myszy, psa i kota - badanie funkcji transformacją Fouriera
Przewidywania dotyczące kształtu trasy psa i kota okazały się prawdziwe - faktycznie przypominały one sinusoidę. Postaramy się jednak dokładniej zbadać ich kształt - wykorzystamy w tym celu transformację Fouriera. Zrobimy to na przykładzie trasy kota.

Transformacja Fouriera pozwala na dekompozycję funkcji na jej składowe częstotliwościowe. Jeśli funkcja jest sinusoidalna, jej transformata Fouriera powinna wykazać wyraźny pik w odpowiadającej jej częstotliwości. Analiza transformacji Fouriera umożliwia również ocenę obecności innych składowych harmonicznych, co może wskazywać na odchylenia od idealnego kształtu sinusoidalnego. Wykorzystanie tego narzędzia pozwoli na dokładne określenie charakterystyki badanej funkcji.

Wydłużymy znacznie czas pogoni za myszą, a do analizy fourierowskiej wykorzystamy część trasy kota dopiero od pewnego momentu w czasie, aby zminimalizować odchylenie spowodowane przez poczatkową trajektorię, która (z powodu pozycji startowej kota) nie przypomina sinusoidy. Sprawdzimy zatem czy jego trasa staje się sinusem po dłuższym czasie gonitwy.  


```python
#mysz, kot i pies
from math import sqrt, cos, sin
import matplotlib.pyplot as plt
import numpy as np
from scipy.fft import fft, fftfreq
import warnings

warnings.filterwarnings("ignore", category=RuntimeWarning)

#((ut1-xt1)*((x1-x0)/dt)+(vt1-yt1)*((yt1-yt0)/dt))/(sqrt((u1t-x1t)**2+(v1t-y1t)**2))-1=0
#((xt1-xt0)/dt)**2+((yt1-yt0)/dt)**2-1=0

from scipy.optimize import fsolve
def przyblizanie(xt0, yt0,ut1,vt1,dt):
    # Funkcja definiująca układ równań
    def equations(vars, xt0, yt0,ut1,vt1,dt):
        xt1, yt1 = vars
        eq1 = ((ut1-xt1)*((xt1-xt0)/dt)+(vt1-yt1)*((yt1-yt0)/dt))/(sqrt((ut1-xt1)**2+(vt1-yt1)**2))-1
        eq2 = ((xt1-xt0)/dt)**2+((yt1-yt0)/dt)**2-1
        return [eq1, eq2]
    # Przybliżenie początkowe
    initial_guess = [xt0+1/10000,yt0+1/10000]

    # Rozwiązywanie układu równań
    solution = fsolve(equations, initial_guess, args=(xt0, yt0,ut1,vt1,dt))

    return solution 
dt=1/100 
u0=0
tabv=[]
tabu=[]
v0=0 
t0=0
tabv.append(v0)
tabu.append(u0)
for i in range (1, 20000):
    v0=(2*sin(6.28*t0))
    tabv.append(v0)
    u0=(t0)
    tabu.append(u0)
    t0=t0+dt

  #  print(-i)

def euler2(tabu2,tabv2,a,b): 
    rozwiazanie=[]
    vt1=tabv2[1]
    ut1=tabu2[1]
    xt0=a 
    yt0=b 
    for i in range (0,len(tabu2)-1):
        para=przyblizanie(xt0, yt0,ut1,vt1,dt)
        rozwiazanie.append(para[0])
        rozwiazanie.append(para[1])
        xt0=para[0]
        yt0=para[1]
        ut1=tabu2[i+1]
        vt1=tabv2[i+1]
       # print(i)
    return(rozwiazanie)

wynik=euler2(tabu,tabv,-4,1) 


x = wynik[::2] 
y = wynik[1::2] 

wynik2=euler2(x,y,-8,-8) 

x2 = wynik2[::2]  
y2 = wynik2[1::2] 



x = x[19000:]
y = y[19000:]

dx = np.diff(x)


fft_y = fft(y)


frequencies = np.fft.fftfreq(len(x), d=np.mean(dx))


plt.plot(frequencies[:len(frequencies) // 2], np.abs(fft_y)[:len(frequencies) // 2])
plt.title("Transformacja Fouriera (Amplituda w dziedzinie częstotliwości)")
plt.xlabel("Częstotliwość [Hz]")
plt.ylabel("Amplituda")
plt.grid(True)

plt.tight_layout()
plt.show()
```

Wynik przeprowadzonej analizy sugeruje, że trasa kota faktycznie zaczęła przypominać sinusoidę w późniejszym etapie gonitwy. Na wykresie transformacji Fouriera można zaobserwować wyraźny pik, który jest charakterystyczny dla funkcji sinusoidalnej. Pik ten występuje w częstotliwości odpowiadającej sinusowi opisującemu trajektorię kota. 

Obecność tego piku potwierdza, że w analizowanym fragmencie ruch kota jest zdominowany przez jedną główną częstotliwość, co wskazuje na oscylacyjny charakter trasy. Dzięki temu można stwierdzić, że trajektoria kota z czasem dąży do sinusoidalnej formy, mimo że początkowo ruch znacząco odbiegał od tego kształtu.

# Przykład - poczwórna pogoń
Drobne poprawki w algorytmie pozwoliły na zasymulowanie pogoni czterech obiektów - niebieski będzie gonił czarnego, czarny czerwonego, czerwony zielonego, a zielony niebieskiego. Intuicja podpowiada, że powstały w ten sposób kształt powinien być "ładny", symetryczny, a punkty powinny się do siebie zbliżać.

```python
#poczworna pogon - ladny wzorek
from math import sqrt, cos, sin
import matplotlib.pyplot as plt
import numpy as np
import warnings

warnings.filterwarnings("ignore", category=RuntimeWarning)



from scipy.optimize import fsolve
def przyblizanie(xt0, yt0,ut1,vt1,dt):
    # Funkcja definiująca układ równań
    def equations(vars, xt0, yt0,ut1,vt1,dt):
        xt1, yt1 = vars
        eq1 = ((ut1-xt1)*((xt1-xt0)/dt)+(vt1-yt1)*((yt1-yt0)/dt))/(sqrt((ut1-xt1)**2+(vt1-yt1)**2))-1
        eq2 = ((xt1-xt0)/dt)**2+((yt1-yt0)/dt)**2-1
        return [eq1, eq2]
    # Przybliżenie początkowe
    initial_guess = [xt0+1/10000,yt0+1/10000]

    # Rozwiązywanie układu równań
    solution = fsolve(equations, initial_guess, args=(xt0, yt0,ut1,vt1,dt))

    return solution 
dt=1/10000 #[s], czas calkowity to 1/1000*10000 czyli 10s

def euler3(x10,y10,x11,y11,x12,y12): # funkcja zalezna od poczatkowych polozen pierszego punktu - nadajemy mu tak jakby poczatkowy kierunek zeby nasza spirala ruszyla
    rozwiazanie1=[] #tablica z punktami pierwszego obiektu
    rozwiazanie2=[] #tablica z punktami drugiego obiektu
    rozwiazanie3=[] #tablica z punktami trzeciego obiektu
    rozwiazanie4=[] #tablica z punktami czwartego obiektu

    rozwiazanie1.append(x10) #dodajemy pierwsze polozenia pierwszego punktu
    rozwiazanie1.append(y10)
    rozwiazanie1.append(x11)
    rozwiazanie1.append(y11)
    rozwiazanie1.append(x12)
    rozwiazanie1.append(y12)

    x20=2 #okreslamy pozycje startowe pozostalych punktow
    y20=2
    x30=2
    y30=-2
    x40=-2
    y40=-2

    for i in range (2,100000,2):
        para2=przyblizanie(x20, y20,rozwiazanie1[i],rozwiazanie1[i+1],dt) #drugi goni pierwszego
        rozwiazanie2.append(para2[0])
        rozwiazanie2.append(para2[1])

        para3=przyblizanie(x30, y30,rozwiazanie2[i-2],rozwiazanie2[i+1-2],dt) #trzeci goni drugiego
        rozwiazanie3.append(para3[0])
        rozwiazanie3.append(para3[1])

        para4=przyblizanie(x40, y40,rozwiazanie3[i-2],rozwiazanie3[i+1-2],dt) #czwarty goni trzeciego
        rozwiazanie4.append(para4[0])
        rozwiazanie4.append(para4[1])

        para1=przyblizanie(x11, y11,rozwiazanie4[i-2],rozwiazanie4[i+1-2],dt) #pierwszy goni czwartego
        rozwiazanie1.append(para1[0])
        rozwiazanie1.append(para1[1])

        x20=para2[0] #podobnie jak w poprzednich algorytmach stawiamy odpowiednie wartosci do nastepnego powtorzenia petli
        y20=para2[1]
        x30=para3[0]
        y30=para3[1]
        x40=para4[0]
        y40=para4[1]
        x11=para1[0]
        y11=para1[1]

        #print(i)

    return(rozwiazanie1, rozwiazanie2, rozwiazanie3, rozwiazanie4) #zwracamy tablice rozwiazan w postaci x1,y1,x2,y2,x3,y3 dla kazdego obiektu

wynik=euler3(-2,2,-1.99,1.99,-1.98,1.98) #wywolujemy funkjce (inicjujemy lekki ruch poczatkowy pierwszego punktu)


x1 = wynik[0][::2]  # Co drugi element zaczynając od 0 to x
y1 = wynik[0][1::2] # Co drugi element zaczynając od 1 to y

x2 = wynik[1][::2]  # Co drugi element zaczynając od 0 to x
y2 = wynik[1][1::2] # Co drugi element zaczynając od 1 to y

x3 = wynik[2][::2]  # Co drugi element zaczynając od 0 to x
y3 = wynik[2][1::2] # Co drugi element zaczynając od 1 to y

x4 = wynik[3][::2]  # Co drugi element zaczynając od 0 to x
y4 = wynik[3][1::2] # Co drugi element zaczynając od 1 to y

# Tworzenie wykresu
plt.figure(figsize=(8, 6))
plt.plot(x1, y1, color='blue', label='trasa pierwszego obiektu') #pierwszy punkt
plt.plot(x2, y2, color='green', label='trasa drugiego obiektu') #drugi punkt
plt.plot(x3, y3, color='red', label='trasa trzeciego obiektu') #trzeci punkt
plt.plot(x4, y4, color='black', label='trasa czwartego obiektu') #czwarty punkt

plt.xlim(-5, 5)  # Zakres osi X 
plt.ylim(-5, 5)  # Zakres osi Y

# Dodanie etykiet i tytułu
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Przykład - poczwórna pogoń')

# Pokazanie wykresu
plt.legend()
plt.grid(True)
plt.show()

```
